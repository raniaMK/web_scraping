# Importing Lib & Methods

```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
import numpy as np
import pandas as pd
```

# Inputing the URL

```python
url = "https://wuzzuf.net/search/jobs/?a=hpb&q=data%20analyst"
```


# create a client-based  Reqest to get the URL

```python
client =urlopen(url)
```

# Getting the HTML code of the full page

```python
html= client.read()
```

```python
html
```

# Closing the request

```python
client.close()
```

# Creating html parser using BeautifulSoup

```python
soup = bs(html,"html.parser")
```

```python
soup
```

# creating a container for the needed Data

```python
containers = soup.find_all("div",{"class":"css-pkv5jc"})
```

```python
len(containers)
```

```python
bs.prettify(containers[0])
```

```python
jobtype = containers[0].find("div", {"class":"css-1lh32fc"})
print(jobtype.text)
```

# Bringing if all Together

```python
file =open("job.csv","w")
header ="job_title/company_name/job_type\n"
print(header)
file.write(header)
```

```python
for container in containers : 
    
    jbt=(container.findAll("h2", {"class":"css-m604qf"}))
    job_title=jbt[0].text.strip()
    
    companyname = container.findAll("a", {"class":"css-17s97q8"})
    company_name=companyname[0].text.strip()
    
    jobtype = container.findAll("div", {"class":"css-1lh32fc"})
    job_type=jobtype[0].text.strip()
    
    print(job_title+ ","+ company_name +","+ job_type)
    
    file.write( job_title+ "/"+ company_name +"/"+ job_type + "\n")
    
    file.close
```
